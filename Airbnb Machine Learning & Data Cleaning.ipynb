```python
dt_records = data.groupby(['country_destination'])
dt_pct = dt_records['country_destination'].count()
total = dt_pct.sum()
dt_pct = dt_pct.apply(lambda x: x * 100 / total)
dt_pct = dt_pct.sort_values(ascending=False)

dt = pd.DataFrame({'Records':data.country_destination.value_counts(), 'Percentage':dt_pct})
dt
```
